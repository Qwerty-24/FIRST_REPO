```python
#109
import math

import math

#sap xep diem theo goc cuc
def order_polygon(points):
    cx = sum(p[0] for p in points) / len(points)
    cy = sum(p[1] for p in points) / len(points)
    return sorted(points, key=lambda p: math.atan2(p[1] - cy, p[0] - cx))

def point_in_quadrilateral(quad, p):
    x, y = p
    
    # neu p trung dinh -> trong bien
    if p in quad:
        return True

    inside = False

    for i in range(4):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % 4]

        # kiem tra nam tren canh
        if ((x2 - x1) * (y - y1) - (y2 - y1) * (x - x1) == 0):
            if min(x1, x2) <= x <= max(x1, x2) and min(y1, y2) <= y <= max(y1, y2):
                return True

        # ray casting
        cut = ((y1 > y) != (y2 > y)) and (x < (x2 - x1) * (y - y1) / (y2 - y1) + x1) 
        # (y1 > y) != (y2 > y) -> p nam giua theo phuong y
        # (x2 - x1) * (y - y1) / (y2 - y1) + x1 -> hoanh do giao diem (neu co) PHAI nam ben phai p
        if cut:
            inside = not inside

    return inside

square = []
for _ in range(4):
    square.append(tuple(map(int, input().split())))

square = order_polygon(square)

p = tuple(map(int, input().split()))

if point_in_quadrilateral(square, p):
    print("not outside")
else:
    print("outside")

# for x in square:
#     print(x)
# print(p)
```
