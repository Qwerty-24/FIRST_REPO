```python
#57
import math

def count_divisor(n):
    cnt = 0
    for i in range(1, int(n**0.5)+1):
        if (n % i == 0):
            cnt += 1
            if not (i == n/i):
                cnt += 1 #tranh truong hop i == sqrt(n)
    return cnt

n = int(input())
# print(count_divisor(n))
print(math.ceil(count_divisor(n)/2))
```

```python
#59
import math

def divisor(n):
    cnt1 = 1
    cnt2 = 1
    for i in range(1, int(n**0.5)+1):
        if (n % i == 0):
            cnt1 = i
            cnt2 = int(n/i)
    print(f"{cnt1} {cnt2}")

n = int(input())
divisor(n)
```
