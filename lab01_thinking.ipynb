### Problem 1
#### Write a function that returns the number that is double the input (e.g., 23 -> 46).

```python
def double_number(x):
    x *= 2
    return x
```

```python
assert double_number(23) == 46
assert double_number(0) == 0
assert double_number(-7) == -14
```

### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def count_in_range(n):
    a = (n - (n-1)//2 - 1)
    return (a)
```

```python
assert count_in_range(9) == 4  # 5,6,7,8
assert count_in_range(2) == 1   # 1 only
assert count_in_range(1) == 0
```

### Problem 5
#### Write a function that returns the product of the integers from 1 to n (factorial). For n=0, return 1.

```python
def factorial(n):
    fac = 1
    for i in range(0, n):
        fac = fac*(i+1)
    return fac
```

```python
assert factorial(0) == 1
assert factorial(5) == 120
assert factorial(1) == 1
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
def triangle_area(a, b, c):
    p = (a + b + c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    return s
```

```python
import math
assert abs(triangle_area(3, 4, 5) - 6.0) < 1e-12
assert abs(triangle_area(2.5, 2.5, 3.0) - 3.0) < 1e-12
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def shift_char(x, n):
    code_x = ord(x)
    code_x += n
    
    if ('A' <= x):
        code_x = ((code_x - ord('A')) % 26 + ord('A'))
    if ('a' <= x):
        code_x = ((code_x - ord('a')) % 26 + ord('a'))
    
    return (chr(code_x))
```

```python
assert shift_char('b', 1) == 'c'
assert shift_char('z', 1) == 'a'
assert shift_char('A', 2) == 'C'
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def to_lower(ch):
    if (ch < 'A' or 'Z' < ch < 'a' or 'z' < ch):
        return ValueError
    if (ch <= 'Z'):
        return chr(ord(ch) + 32)
    else: return ch
```

```python
assert to_lower('A') == 'a'
assert to_lower('Z') == 'z'
assert to_lower('a') == 'a'
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
def repeat_lower(ch):
    s = ch*10
    return s
```

```python
assert repeat_lower('a') == 'aaaaaaaaaa'
assert repeat_lower('z') == 'zzzzzzzzzz'
assert len(repeat_lower('m')) == 10
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def right_isosceles(n):
    s = ""
    for i in range(0,n):
        s += '*' * (i+1)
        s += "\n"
    return s[:-1] if s else ""
```

```python
assert right_isosceles(4) == '*\n**\n***\n****'
assert right_isosceles(1) == '*'
assert right_isosceles(0) == ''
```
