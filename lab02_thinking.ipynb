### Problem 1

**Task:** Write a function `is_product_divisible_by_10(a, b)` that returns `True` if `a * b` is divisible by 10, otherwise `False`.
- Inputs: integers `a`, `b` (can be negative or zero).
- Output: boolean.
- Hint: A number is divisible by 10 if it ends with digit 0 (i.e., divisible by both 2 and 5).


```python

def is_product_divisible_by_10(a: int, b: int) -> bool:
    if (a*b % 10 == 0): return True
    else: return False
```

```python

assert is_product_divisible_by_10(0, 7) is True        # 0 is divisible by 10
assert is_product_divisible_by_10(5, 2) is True        # 10 divisible by 10
assert is_product_divisible_by_10(3, 10) is True
assert is_product_divisible_by_10(-4, 5) is True
assert is_product_divisible_by_10(7, 11) is False
assert is_product_divisible_by_10(-3, -7) is False
print("Problem 1 tests passed!")
```

### Problem 2

**Task:** Write a function `divide_with_remainder(dividend, divisor)` that returns a tuple `(q, r)` where `q` is the integer quotient and `r` is the remainder such that `dividend = divisor*q + r` with `0 <= |r| < |divisor|` (use Python's `//` and `%` semantics).
- Inputs: integers `dividend`, `divisor` (`divisor` must not be 0).
- Output: tuple of two integers `(q, r)`.
- Example: `divide_with_remainder(5, 3) -> (1, 2)`.


```python

def divide_with_remainder(dividend: int, divisor: int) -> tuple[int, int]:
    return (dividend // divisor, dividend % divisor)
```

```python
assert divide_with_remainder(5, 3) == (1, 2)
assert divide_with_remainder(5, -3) == (-2, -1)
assert divide_with_remainder(-5, 3) == (-2, 1)
assert divide_with_remainder(0, 7) == (0, 0)
assert divide_with_remainder(100, 10) == (10, 0)
print("Problem 2 tests passed!")
```

### Problem 3

**Task:** Write a function `max_of_two(a, b)` that returns the larger number.
- Inputs: numbers `a`, `b` (ints or floats).
- Output: the maximum of `a` and `b`.


```python

def max_of_two(a, b):
    if (a > b): return a
    return b
```

### Problem 4

**Task:** Write a function `compute(a, b, op)` that computes `a (+|-|*|/) b` depending on `op`.
- Inputs: numbers `a`, `b` and a character `op` in `{'+', '-', '*', '/'}`.
- Output: the numeric result.
- Requirements: 
  - If `op` is `'/'` and `b == 0`, raise `ValueError`.
  - If `op` is invalid, raise `ValueError`.


```python
def compute(a, b, op):
    ope = {
        '+': (a + b),
        '-': (a - b),
        '*': (a * b),
        '/': (a / b) if (b!=0) else ValueError 
    }
    return ope.get(op, ValueError)

# a, b = map(int, input().split())
# op = input()
# print(compute(a, b, op))
```

- Inputs: positive numbers `a`, `b`, `c`.
- Output: boolean.
- Notes: Check triangle inequality and the Pythagorean theorem. Use a small tolerance for floating-point comparisons.


```python
import math
def is_right_triangle(a: float, b: float, c: float) -> bool:
    if (a <= 0 or b <= 0 or c <= 0): return False
    # if (a**2+b**2 == c**2 or b**2+c**2 == a**2 or c**2 + a**2 == b**2): return True
    # else: return False
    # doi khi khong chinh xac do loi lam tron -> math.isclose
    return(math.isclose(a**2+b**2, c**2) or math.isclose(a**2+b**2, c**2) or math.isclose(c**2 + a**2, b**2))


```

### Problem 6

**Task:** Write a function `is_perfect_square(n)` that returns `True` if `n` is a perfect square, otherwise `False`.
- Inputs: integer `n` (can be negative).
- Output: boolean.
- Note: Negative numbers are not perfect squares in the set of integers.


```python
def is_perfect_square(n: int) -> bool:
    k = n**0.5
    if (n == k*k): return True
    return False
```

```python
assert is_perfect_square(0) is True
assert is_perfect_square(1) is True
assert is_perfect_square(16) is True
assert is_perfect_square(15) is False
assert is_perfect_square(-4) is False
assert is_perfect_square(10**12) is True   # (10^6)^2
print("Problem 6 tests passed!")
```

### Problem 7

**Task:** Write a function `detect_number_type(s)` that returns `'int'` if `s` represents an integer, `'float'` if it represents a floating-point number, otherwise `'neither'`.
- Inputs: string `s`.
- Output: one of `{'int', 'float', 'neither'}`.
- Rules:
  - Allow optional leading `+` or `-`.
  - Integers must be digits without a decimal point (e.g., `'100'`).
  - Floats must contain a decimal point with digits on **both** sides (e.g., `'100.0'`, `'-3.14'`).
  - Strings like `'.5'`, `'5.'`, `'1e3'` should be `'neither'` for this task.


```python
def detect_number_type(s: str) -> str:
    for i in s:
        if (i not in ".+-0123456789"): return "neither"
        
    try: 
        int(s)
        return "int"
    except ValueError:
        a, b = map(str, s.split('.'))
        if (a == "" or b == ""): return "neither"
        return "float"

```

```python

# Tests
assert detect_number_type('100') == 'int'
assert detect_number_type('+100') == 'int'
assert detect_number_type('100.0') == 'float'
assert detect_number_type('-3.14') == 'float'
assert detect_number_type('1e3') == 'neither'
assert detect_number_type('.5') == 'neither'
assert detect_number_type('5.') == 'neither'
assert detect_number_type('abc') == 'neither'
print("Problem 7 tests passed!")
```

### Problem 8

**Task:** Write a function `is_one_product_of_others(a, b, c)` that returns `True` if one of the three numbers equals the product of the other two; otherwise `False`.
- Inputs: numbers `a`, `b`, `c` (ints or floats).
- Output: boolean.
- Consider corner cases like zeros and negatives.


```python
def is_one_product_of_others(a, b, c):
    if (a == b*c or b == c*a or c == a*b): return True
    return False
```

```python

# Tests
assert is_one_product_of_others(6, 2, 3) is True     # 6 == 2*3
assert is_one_product_of_others(2, 6, 3) is True     # 6 == 2*3
assert is_one_product_of_others(2, 3, 7) is False
assert is_one_product_of_others(0, 0, 5) is True     # 0 == 0*5
assert is_one_product_of_others(0, 2, 0) is True     # 0 == 0*2
assert is_one_product_of_others(-6, -2, 3) is True   # -6 == -2*3
print("Problem 8 tests passed!")
```

### Problem 9

**Task:** Write a function `sum_halves(s)` that takes a string `s` of exactly 6 digits and returns the sum of the first three-digit integer and the last three-digit integer.
- Inputs: string `s` of length 6, consisting only of digits 0-9.
- Output: integer sum.
- Example: `'100200' -> 100 + 200 = 300`.
- If input is invalid (wrong length or contains non-digits), raise `ValueError`.


```python

def sum_halves(s: str) -> int:
    if (s.isdigit() == False or len(s) !=  6): return ValueError
    return int(s[:3]) + int(s[3:])

```

```python

# Tests
assert sum_halves('100200') == 300
assert sum_halves('000000') == 0
assert sum_halves('999999') == 1998
assert sum_halves('001234') == (1 + 234)
print("Problem 9 tests passed!")
```

### Problem 10

**Task:** Write a function `quadratic_roots_count(a, b, c)` that returns the number of **real** solutions of `a*x^2 + b*x + c = 0`.
- Inputs: real numbers `a`, `b`, `c`.
- Output: `0`, `1`, or `2`. If the equation degenerates to `0 = 0` (i.e., `a == 0`, `b == 0`, `c == 0`), return `float('inf')`.
- Rules:
  - If `a != 0`, use the discriminant `D = b^2 - 4ac`.
  - If `a == 0` and `b != 0`, it's linear with exactly one real root.
  - If `a == 0` and `b == 0` and `c != 0`, there is no solution.


```python

def quadratic_roots_count(a: float, b: float, c: float):
    if (a == 0):
        if (b != 0): return 1
        elif (c == 0): return float('inf')
        return 0
    else:
        if (b*b - 4*a*c == 0): return 1
        elif (b*b - 4*a*c > 0): return 2
        return 0

    
```

```python

# Tests
import math
assert quadratic_roots_count(1, 0, -1) == 2       # x^2 - 1 = 0
assert quadratic_roots_count(1, 2, 1) == 1        # (x+1)^2 = 0
assert quadratic_roots_count(1, 0, 1) == 0        # x^2 + 1 = 0
assert quadratic_roots_count(0, 2, 1) == 1        # 2x + 1 = 0
assert quadratic_roots_count(0, 0, 1) == 0        # no solution
assert math.isinf(quadratic_roots_count(0, 0, 0)) # infinite solutions
print("Problem 10 tests passed!")
```

### Problem 12

**Task:** Write a function `point_in_quadrilateral(quad, p)` that returns `True` if point `p` lies **inside or on the boundary** of the quadrilateral `quad`, and `False` otherwise.
- Inputs: 
  - `quad`: a list or tuple of exactly 4 vertices `[(x1,y1), (x2,y2), (x3,y3), (x4,y4)]` in order (the quadrilateral may be convex or concave, non-self-intersecting).
  - `p`: a point `(x, y)`.
- Output: boolean.
- Hint: Implement a general point-in-polygon test (e.g., ray casting), treating boundary as inside.


```python
import math
def point_in_quadrilateral(quad: list[tuple[float, float]] | tuple[tuple[float, float], ...], 
                           p: tuple[float, float]) -> bool:
    if (tri_area(*quad[0], *quad[1], *quad[2]) + tri_area(*quad[0], *quad[2], *quad[3]) == tri_area(*quad[0], *quad[1], *p) + tri_area(*quad[1], *quad[2], *p) + tri_area(*quad[2], *quad[3], *p) + tri_area(*quad[3], *quad[0], *p)): return True
    return False

def tri_area(x1, y1, x2, y2, x3,  y3):
    return (1/2)*abs(x1*(y2 - y3) + x2*(y3 - y1) +  x3*(y1- y2))
```

```python
import math

def tri_area(x1, y1, x2, y2, x3, y3):
    return 0.5 * abs(
        x1 * (y2 - y3) +
        x2 * (y3 - y1) +
        x3 * (y1 - y2)
    )

def point_in_quadrilateral(
    quad: list[tuple[float, float]] | tuple[tuple[float, float], ...],
    p: tuple[float, float]
) -> bool:
    # diện tích tứ giác chia thành 2 tam giác
    quad_area = (
        tri_area(*quad[0], *quad[1], *quad[2]) +
        tri_area(*quad[0], *quad[2], *quad[3])
    )
    
    # tổng diện tích của 4 tam giác với p
    sum_area = (
        tri_area(*quad[0], *quad[1], *p) +
        tri_area(*quad[1], *quad[2], *p) +
        tri_area(*quad[2], *quad[3], *p) +
        tri_area(*quad[3], *quad[0], *p)
    )
    
    return math.isclose(quad_area, sum_area, rel_tol=1e-9, abs_tol=1e-9)
```

```python

# Tests
# Convex square
square = [(0,0), (2,0), (2,2), (0,2)]
assert point_in_quadrilateral(square, (1,1)) is True     # inside
assert point_in_quadrilateral(square, (0,1)) is True     # on boundary
assert point_in_quadrilateral(square, (2,2)) is True     # vertex
assert point_in_quadrilateral(square, (-0.1,1)) is False # outside
assert point_in_quadrilateral(square, (3,3)) is False    # far outside

# Concave quad (arrow shape)
concave = [(0,0), (3,1), (0,2), (1,1)]
assert point_in_quadrilateral(concave, (1,1)) is True    # vertex (boundary)
assert point_in_quadrilateral(concave, (1.2,1)) is True  # inside concave area
assert point_in_quadrilateral(concave, (2.5,1.5)) is False
print("Problem 12 tests passed!")
```
