### Problem 7

#### Write a function that returns the number of non-zero digits in n.

```python
def count_nonzero_digits(n):
    n = abs(n)
    n = str(n).replace('0', '')
    return len(n)
```

### Problem 9

#### Write a function that returns the angle ∠ABC in degrees given coordinates of points A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def angle_ABC(x1, y1, x2, y2, x3, y3):
    # Vector BA và BC
    BAx, BAy = x1 - x2, y1 - y2
    BCx, BCy = x3 - x2, y3 - y2

    # Tích vô hướng
    dot = BAx * BCx + BAy * BCy

    # Độ dài 2 vector
    normBA = math.sqrt(BAx**2 + BAy**2)
    normBC = math.sqrt(BCx**2 + BCy**2)

    # Tránh chia cho 0
    if normBA == 0 or normBC == 0:
        raise ValueError("Các điểm không hợp lệ (B trùng A hoặc C).")

    # cos(theta)
    cos_theta = dot / (normBA * normBC)

    # Giới hạn cos_theta trong [-1, 1] để tránh lỗi do sai số số học
    cos_theta = max(-1, min(1, cos_theta))

    # Tính góc (radian -> độ)
    theta = math.degrees(math.acos(cos_theta))
    return theta
```

```python
def angle_ABC(x1, y1, x2, y2, x3, y3):
    xAB = x1 - x2
    yAB = y1 - y2
    xCB = x3 - x2
    yCB = y3 - y2
    if (xAB*xCB + yAB*yCB == 0): return 90
    return 180
```

```python
assert round(angle_ABC(0, 0, 0, 1, 1, 1), 5) == 90.00000
assert round(angle_ABC(0, 0, 1, 0, 2, 0), 5) == 180.00000
assert round(angle_ABC(1, 0, 0, 0, 0, 1), 5) == 90.00000
assert round(angle_ABC(0, 0, 1, 1, 2, 2), 5) == 180.00000
assert round(angle_ABC(1, 0, 0, 0, -1, 0), 5) == 180.00000
```

#### Write a function that returns the maximum and minimum of three numbers a, b, c as a list [max, min]. (No tuples or dictionaries)

```python
def max_min(a, b, c):
    l = []
    if (b < c): 
        b, c = c, b # b >= c
    if (a < c):
        l += [b, a]
    elif (a < b):
        l += [b, c]
    else:
        l += [a, c]
    return l
```

### Problem 11

#### Write a function that returns the binomial coefficient C(n, k).

```python
# n*(n-1)*(n-2)*...*(n-k+1)/ 1*2*3*...*k

# 1*2*...*n/ 1*2*...*k * 1*2*...*(n-k)

def combination(n, k):
    k = min(k, n - k) # nCk = nC(n-k)
    res = 1
    for i in range (1, k+1):
        res *= (n-i+1)
        res /= i
    
    return res

# import math
# def combination(n, k):
#     return (math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))



```

### Problem 12

#### Write a function that returns the sum of squares of numbers from 1 to n.

```python
def sum_squares(n):
    s = 0
    for i in range(1, n+1):
        s += i*i
    return s
```

```python
assert sum_squares(1) == 1
assert sum_squares(2) == 5
assert sum_squares(3) == 14
assert sum_squares(5) == 55
assert sum_squares(10) == 385
```
