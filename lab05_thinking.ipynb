### Problem 1

#### Write a function `sum_1_to_n(n)` that **returns** the sum of all integers from 1 to `n`.  If `n` is non-positive, return 0.

```python
def sum_1_to_n(n):
    if (n < 0): return 0
    sum = 0;
    for i in range(1, n+1):
        sum += i
    return sum
```

```python
# Boundary and typical tests
assert sum_1_to_n(0) == 0
assert sum_1_to_n(-5) == 0
assert sum_1_to_n(1) == 1
assert sum_1_to_n(5) == 15
assert sum_1_to_n(10) == 55
```

### Problem 2

#### Write a function `sum_even_1_to_n(n)` that returns the sum of all even integers from 1 to `n` inclusive. If `n` is non-positive, return 0.

```python
def sum_even_1_to_n(n):
    if (n < 0): return 0
    sum = 0
    for i in range(1, n+1):
        if (i & 1 == 0): # if not (i & 1)
            sum += i
    return sum
```

```python
assert sum_even_1_to_n(0) == 0
assert sum_even_1_to_n(1) == 0
assert sum_even_1_to_n(2) == 2
assert sum_even_1_to_n(6) == 2 + 4 + 6
assert sum_even_1_to_n(7) == 2 + 4 + 6
```

### Problem 4

#### Write a function `sum_multiples_upto_n(m, n)` that **returns** the sum of all numbers from 1 to `n` that are divisible by `m`.  If `n` is non-positive or `m` is 0, return 0.

```python
def sum_multiples_upto_n(m, n):
    if (m <= 0 or n <= 0): return 0
    sum = 0
    for i in range(1, n+1):
        if (i % m == 0):
            sum += i
    return sum
```

```python
# Boundary and typical tests
assert sum_multiples_upto_n(0, 10) == 0
assert sum_multiples_upto_n(3, 0) == 0
assert sum_multiples_upto_n(3, 10) == 18   # 3 + 6 + 9
assert sum_multiples_upto_n(2, 1) == 0
assert sum_multiples_upto_n(5, 25) == 75   # 5 + 10 + 15 + 20 + 25
```

### Problem 5

#### Write a function `square_matrix_of_n(n)` that **returns** a string representing an `n x n` matrix filled with the number `n`. Each row uses a single space between elements, and rows are separated by newline characters. If `n` is non-positive, return an empty string.

```python
def square_matrix_of_n(n):
    s = ""
    for i in range (1, n+1):
        s += (f"{n} " * n).strip()
        s += "\n"

    s = s.rstrip()
    return s
```

```python
assert square_matrix_of_n(0) == ""
assert square_matrix_of_n(-3) == ""
assert square_matrix_of_n(1) == "1"
assert square_matrix_of_n(2) == "2 2\n2 2"
assert square_matrix_of_n(3) == "3 3 3\n3 3 3\n3 3 3"
```

### Problem 6

#### Write a function `rectangle_of_char(n, m, c)` that **returns** a string representing a grid with `m` rows and `n` columns, containing the character `c`. Elements in each row are separated by a single space, and rows are separated by newline characters. If `n <= 0` or `m <= 0`, return an empty string.

```python
def rectangle_of_char(n, m, ch):
    if (n <= 0 or m <= 0): return ""
    s = ""
    for i in range (1, m+1):
        s += (f"{ch} " * n).strip()
        s += "\n"

    s = s.rstrip()
    return s
```

```python
assert rectangle_of_char(0, 3, "#") == ""
assert rectangle_of_char(3, 0, "#") == ""
assert rectangle_of_char(1, 1, "@") == "@"
assert rectangle_of_char(2, 3, "*") == "* *\n* *\n* *"
assert rectangle_of_char(4, 2, "X") == "X X X X\nX X X X"
```

### Problem 7  

#### Write a function `inverted_triangle(n)` that returns a centered inverted triangle of asterisks with width `n`.  
- `n` must be a positive odd integer, otherwise return an empty string.  
- Each line has exactly `n` characters, decreasing by 2 asterisks per line until one remains.  

Example `n = 5` in test cases


```python
def inverted_triangle(n):
    if not (n & 1): return ""
    slist = []

    for i in range(n+1, 1, -1):
        if (i & 1):
            s = ""
            k = ((n-i)/2)
            s += ' ' * k + '*'*i + ' ' * k + "\n"
            slist.append(s)
    
    for i in slist:
        print(i)
    
```

```python
tri5 = inverted_triangle(5).split("\n")
assert tri5[0] == "*****"
assert tri5[1] == " *** "
assert tri5[2] == "  *  "

assert inverted_triangle(1) == "*"
assert inverted_triangle(3) == "***\n * "
assert inverted_triangle(7) == "*******\n ***** \n  ***  \n   *   "
```
