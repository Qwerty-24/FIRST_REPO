### Problem 12

**Task:** A string is **perfect** if it contains at least one uppercase letter, at least one lowercase letter, and at least one digit.  
Return `True` if the string is perfect; otherwise return `False`. If input is not a string, convert it.

```python
def is_perfect_string(s):
    cup = 0
    clow = 0
    cdi = 0
    for i in range(0, len(s)):
        if (s[i].isupper()): cup = 1
        if (s[i].islower()): clow = 1
        if (s[i].isdigit()): cdi = 1
    
    if (cup and clow and cdi): return True
    else: return False
```

```python
assert is_perfect_string("Aa0") is True
assert is_perfect_string("abc") is False
assert is_perfect_string("ABC") is False
assert is_perfect_string("123") is False
assert is_perfect_string("Passw0rd") is True
```
