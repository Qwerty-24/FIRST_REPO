```python
x, n = map(str, input().split())
n = int(n)

def shift_char(x, n):
    code_x = ord(x)
    code_x += n
    
    if ('A' <= x):
        code_x = ((code_x - ord('A')) % 26 + ord('A'))
    if ('a' <= x):
        code_x = ((code_x - ord('a')) % 26 + ord('a'))
    
    print(chr(code_x))

shift_char(x, n)
```

```python
n, m = map(int, input().split())

def rectangle(n, m):
    recta = []
    for i in range (0, n):
        recta.append('*' * (m))
    return '\n'.join(recta)

print(rectangle(2,3))
```
