```python
x, n = map(str, input().split())
n = int(n)

def shift_char(x, n):
    code_x = ord(x)
    code_x += n
    
    if ('A' <= x):
        code_x = ((code_x - ord('A')) % 26 + ord('A'))
    if ('a' <= x):
        code_x = ((code_x - ord('a')) % 26 + ord('a'))
    
    print(chr(code_x))

shift_char(x, n)
```

```python
data = {"a": 1, "b": 2, "c": 3}

# Lấy danh sách keys
print(data.keys())  # dict_keys(['a', 'b', 'c'])

# Lấy danh sách values
print(data.values())  # dict_values([1, 2, 3])

# Lấy danh sách cặp key-value
print(data.items())  # dict_items([('a', 1), ('b', 2), ('c', 3)])

# Kiểm tra key có tồn tại không
print("a" in data)  # True
print("z" in data)  # False

# Xóa key
data.pop("b")
print(data)  # {'a': 1, 'c': 3}

# Lấy value với default nếu key không tồn tại
print(data.get("b", 0))  # 0
```

```python
def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

print(is_integer("-320"))   # True
print(is_integer("3a7"))   # False
print(is_integer("1.0"))   # True
  # False
```
